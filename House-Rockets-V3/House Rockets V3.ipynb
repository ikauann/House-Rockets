```python
import pandas as pd

pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

# Qual o número e imóveis por ano de construção ?

```python
df[['id', 'yr_built']].groupby('yr_built').count()#groupby agrupa segundo a uma informação.
```

# Qual menor número de quartos por ano de construção de imóveis ?

```python
df[['bedrooms', 'yr_built']].groupby('yr_built').min()#groupby agrupa segundo a uma informação.
```

# Qual o preço de compra mais alto por cada número de quarto ?

```python
df[['price', 'bedrooms']].groupby('bedrooms').max()
```

# Qual a soma de todos os preços de compra por número de quartos ?

```python
df[['price', 'bedrooms']].groupby('bedrooms').sum()
```

# Qual a soma de todos os preços de compra por número de quartos e banheiros ?

```python
df[['price', 'bedrooms', 'bathrooms']].groupby(['bedrooms', 'bathrooms']).sum()
```

# Qual o tamanho médio das salas dos imóveis por ano de construção ?

```python
df[['sqft_living', 'yr_built']].groupby('yr_built').mean()
```

# Qual o tamanho mediano das sala dos imóveis por ano de construção ?

```python
df[['sqft_living', 'yr_built']].groupby('yr_built').median()
```

# Qual o desvio-padrão do tamanho das salas dos imóveis por ano de construção ?

```python
df[['sqft_living', 'yr_built']].groupby('yr_built').std()
```

# Como é o crescimento médio do preço de compras dos imóveis, por ano, por dia e pela semana do ano ?

###### obs: usando graficos

##### Por ano:

```python
from matplotlib import pyplot as plt

df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
df['ano'] = pd.to_datetime(df.date).dt.year#rodar so uma vez, se nao da erro.
por_ano = df[['price', 'ano']].groupby('ano').sum().reset_index()

plt.figure(figsize=(20, 12))
plt.bar(por_ano['ano'], por_ano['price'])
```

##### Por dia:

```python
df['dia'] = pd.to_datetime(df.date)
por_dia = df[['price', 'dia']].groupby('dia').mean().reset_index()

plt.figure(figsize=(20, 12))
plt.plot(por_dia['dia'], por_dia['price'])
```

##### Por semana:

```python
df['semana'] = pd.to_datetime(df.date).dt.strftime('%Y-%U')
por_semana = df[['price', 'semana']].groupby('semana').mean().reset_index()

plt.figure(figsize=(20, 12))
plt.plot(por_semana['semana'], por_semana['price'])
plt.xticks(rotation=50);
```

```python
import plotly.express as px

house = df[['id', 'lat', 'long', 'price']]

fig = px.scatter_mapbox(

                    house,
                    lat = 'lat',
                    lon = 'long',
                    size = 'price',
                    color_continuous_scale=px.colors.cyclical.IceFire,
                    size_max=15,
                    zoom=10
    
    )

fig.update_layout(mapbox_style='open-street-map')
fig.show()
```
