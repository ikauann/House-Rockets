# Criar uma classfificação para o imóveis, separando-os em baixo e alto padrão, de acordo com o preço.

```python
alto_nivel = df.query('price > 540000')
alto_nivel['level'] = 'alto preco'

baixo_nivel = df.query('price < 540000')
baixo_nivel['level'] = 'baixo preco'

df = pd.DataFrame()
df = df.append(alto_nivel)
df = df.append(baixo_nivel)
df.reset_index(inplace=True)

del df['level_0']
del df['index']

df
```

# Relatorio de id, data que o imovel ficou disponivel para compra, numero de quartos, tamanho total do terreno, preço e a classficação do imovel

```python
df.columns
```

```python
relatorio = df[['id','date', 'price', 'bedrooms', 'sqft_lot', 'level', 'lat', 'long']]
relatorio.sort_values('price')

relatorio.reset_index(inplace=True)

del relatorio['level_0']
del relatorio['index']

relatorio
```

# mapa indicando onde esta localizado o terreno

```python
relatorio.query('price > 5400000').max()
```
